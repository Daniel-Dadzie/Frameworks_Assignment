# ## ðŸ§¹ Part 2: Data Cleaning and Preparation

```python
# Drop columns with >70% missing values
df = df.loc[:, df.isnull().mean() < 0.7]
```

```python
print(f"After dropping columns with >70% missing values:")
print(df.columns)
```

```python
print(df.isnull().sum().head(100))
```

```python
## fill columns with less missing numbers
fill_values = {col: ('Unknown' if df[col].dtype == 'object' else df[col].median()) for col in df.columns}
## fill columns with less missing numbers
fill_values = {}
for col in df.columns:
    if df[col].dtype == 'object':
        fill_values[col] = 'Unknown'
    elif pd.api.types.is_numeric_dtype(df[col].dtype):
        fill_values[col] = df[col].median()
    else:
        fill_values[col] = df[col].mode().iloc[0] if not df[col].mode().empty else 'Unknown'
df.fillna(fill_values, inplace=True)
```

```python
print(f"After filling values with less missing numbers")
print(df.isnull().sum().head(100))
```

```python
# convert date columns to datetime format
# Convert publish_time to datetime and extract year
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
df['year'] = df['publish_time'].dt.year
```

```python
# Abstract word count
df['abstract'] = df['abstract'].fillna('').astype(str)
df['abstract_word_count'] = df['abstract'].str.findall(r'\w+').str.len()
df['abstract_word_count'] = df['abstract'].fillna('').apply(lambda x: len(x.split()))
```


## ðŸ“Š Part 3: Data Analysis and Visualization


```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import re
```

```python
# Publications by year
year_counts = cleaned_data['year'].value_counts().sort_index()
plt.bar(year_counts.index, year_counts.values)
plt.title('Publications by Time')
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.show()
```

```python
# Top 10 Journals
top_journals = cleaned_data['journal'].value_counts().head(10)
top_journals.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Journals')
plt.ylabel('Number of Publications')
plt.show()
```

```python


# Word frequency in titles
import re
from collections import Counter
titles = ' '.join(cleaned_data['title'].dropna().astype(str)).lower()
words = re.findall(r'\b[a-z]{4,}\b', titles)
common_words = Counter(words).most_common(20)
common_words
```

```python

%%writefile app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

cleaned_data = pd.read_csv('metadata.csv', low_memory=False, nrows=100)
cleaned_data['publish_time'] = pd.to_datetime(cleaned_data['publish_time'], errors='coerce')
cleaned_data['year'] = cleaned_data['publish_time'].dt.year

st.title("CORD-19 Data Explorer")
st.write("Explore COVID-19 research metadata interactively.")

years = sorted(cleaned_data['year'].dropna().unique())
year_range = st.slider("Select Year Range", int(min(years)), int(max(years)), (2020, 2021))
filtered_df = cleaned_data[(cleaned_data['year'] >= year_range[0]) & (cleaned_data['year'] <= year_range[1])]

st.subheader("Publications by Year")
year_counts = filtered_df['year'].value_counts().sort_index()
fig, ax = plt.subplots()
ax.bar(year_counts.index, year_counts.values)
st.pyplot(fig)

st.subheader("Top Journals")
top_journals = filtered_df['journal'].value_counts().head(10)
st.bar_chart(top_journals)

st.subheader("Word Cloud of Titles")
titles = " ".join(filtered_df['title'].dropna().astype(str))
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(titles)
st.image(wordcloud.to_array())

st.subheader("Sample Data")
st.dataframe(filtered_df.head(20))
```
